```python
import pandas as pd

dataset=pd.read_csv("Social_Network_Ads.csv")

dataset.head()
```

```python
d1=dataset.iloc[:,0]
d2=dataset.iloc[:,1]

print(d1.head())
print(d2.head())
```

### **MIN-MAX NORMALIZATION**

```python
newmax=max(d1)
newmin=min(d1)

nage=[]
nsalary=[]

for a in d1:
    newwbc=(a-newmin)/(newmax-newmin)
    nage.append(newwbc)

for a in d2:
    newne=(a-newmin)/(newmax-newmin)
    nsalary.append(newne)
```

```python
import matplotlib.pyplot as plt

dataset["new_age"]=nage
dataset["new_avgsal"]=nsalary

plt.scatter(nage,nsalary)
plt.show()

plt.scatter(d1,d2)
plt.show()

print(dataset["new_age"].mean())
print(dataset["new_age"].std())

print(dataset["new_avgsal"].mean())
print(dataset["new_avgsal"].std())
```

### **Z-SCORE NORMALIZATION**

```python
mean=d1.mean()
stdev=d1.std()

mean1=d2.mean()
stdev1=d2.std()

z1=[]
z2=[]

for a in d1:
  z1.append((a-mean)/stdev)

for a in d2:
  z2.append((a-mean1)/stdev1)
```

```python
dataset["newzscoreage"]=z1
dataset["newzscoresalary"]=z2

plt.scatter(z1,z2)
plt.show()

plt.scatter(d1,d2)
plt.show()

print(dataset["newzscoreage"].mean())
print(dataset["newzscoreage"].std())

print(dataset["newzscoresalary"].mean())
print(dataset["newzscoresalary"].std())
```
