<h3>AND and OR Gate using NEURAL NETWORK with PERCEPTRON LEARNING</h3>

```python
import numpy as np
def activation(out,threshold):
    if out>=threshold:
        return 1
    else:
        return 0
```

<h4>AND GATE</h4>

```python
def perceptron(and_input):
    a=[0,0,1,1]
    b=[0,1,0,1]
    y=[0,0,0,1] #Actual ouput of AND gate
    w=[0.1,0.8]
    bias=0.4
    threshold=1
    l_rate=0.5
    i=0
    while (i<4):
        summation=a[i]*w[0]+b[i]*w[1]+bias
        o=activation(summation,threshold)
        print("Input:  " + str(a[i])+  " , " +str(b[i] ))
        print("Weights:  " + str(w[0])+  " , " +str(w[1] ))
        print("Summation:  " + str(summation))
        print("Threshold: " +str(threshold))
        print("Actual Output:  "+ str(y[i]))
        print("Predicted Output: " +str(o))

        if(o != y[i]):
            w[0]=w[0]+l_rate*(y[i]-o)*a[i]
            w[1]=w[1]+l_rate*(y[i]-o)*b[i]
            bias=bias+l_rate*(y[i]-o)
            
            print("Updated Weights:  "+ str(w[0])+  " , " +str(w[1] ))
            print("Updated bias: "+ str(bias))
            i = -1
            print("\nWeights Updated")
        i+=1
  
    #prediction
    summation=and_input[0]*w[0]+and_input[1]*w[1]+bias
    return activation(summation,threshold)

and_input=[0,1]
print("AND Gate Output for" +str(and_input)+ ":" +str(perceptron(and_input)))
```

<h4>OR GATE</h4>
