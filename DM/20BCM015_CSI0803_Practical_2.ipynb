```python
import pandas as pd

dataset=pd.read_csv("city_new.csv")

dataset.head()
```

```python
a1 = ((1-(dataset.isnull().sum()/dataset.shape[0]))*100)
print(a1)
# PRINTS PERCENTAGE OF DATA THAT IS NOT NULL
```

```python
datanew = dataset.dropna(axis=1,thresh=(0.65*dataset.shape[0]))
print(((1-(datanew.isnull().sum()/dataset.shape[0]))*100))
# COLUMN DROPPED FOR COLUMN WITH EMPTY DATA PERCENTAGE MORE THAN 35%
```

<h3>DATA FILLING USING MEAN</h3>

```python
datanew.mean(numeric_only=True)
datanew1 = datanew.fillna(value=dataset.mean())
datanew1.to_csv("city_new1.csv")
```

<h3>DATA FILLING USING MODE OF DATA</h3>

```python
print(datanew['NO'].mode())
datanew2 = datanew
datanew2['NO'].fillna(datanew['NO'].mode()[0], inplace=True)
datanew2.to_csv("city_new2.csv")
datanew2.iloc[:10,:]
```

<H3>DATA FILLING USING FORWARD FILL</H3>

```python
datanew3 = datanew
datanew3['NO2'].fillna(method='ffill',inplace=True)
datanew3.iloc[:59,:]
```

<h3>DATA FILLING USING BACKWARD FILL</h3>

```python
datanew4 = datanew
datanew4['NO2'].fillna(method='bfill',inplace=True)
datanew4.iloc[:59,:]
```

<h3>DATA FILLING USING A CONSTANT NUMBER</h3>

```python
datanew5 = datanew
datanew5['PM2.5'].fillna(0,inplace=True)
datanew5
```

<h3>DATA FILLING USING INTERPOLATION METHOD [LINEAR]</h3>

```python
datanew6 = datanew
datanew6.interpolate(method='linear')
datanew6
```

<h3>DATA FILLING USING INTERPOLATION METHOD [POLYNOMIAL][ORDER 2]</h3>

```python
datanew7 = datanew
datanew7.interpolate(method='polynomial',order=2,inplace=True)
print(datanew7)
datanew.to_csv("a.csv")
datanew7.to_csv("b.csv")
print(datanew)
```

<H3>DATA FILLING USING INTERPOLATION METHOD [POLYNOMIAL][ORDER 3]</H3>

```python
datanew8 = datanew
datanew8.interpolate(method='polynomial',order=3)
datanew8
```
